```
def generate_square_polygons(lon_min, lon_max, lat_min, lat_max, step):
    polygons = []
    lon = lon_max
    while lon > lon_min:
        lat = lat_max
        while lat > lat_min:
            square = [
                [lon, lat],
                [lon - step, lat],
                [lon - step, lat - step],
                [lon, lat - step],
                [lon, lat]
            ]
            polygons.append(square)
            lat -= step
        lon -= step
    return polygons

# Define the coordinate range and step size
# (Contiguous United States)
lon_min = -125.0
lon_max = -66.0
lat_min = 25.0
lat_max = 50.0
step = 0.5

# Generate the list of square polygons
squares = generate_square_polygons(lon_min, lon_max, lat_min, lat_max, step)
```
