```
import requests
from bs4 import BeautifulSoup

def scrape_price_from_website(url):
    # Send a GET request to the website
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find the element with the specified attribute
        price_element = soup.find_all(class_="sc-167fkr-0 XmvAr")

        # Check if the element was found
        if price_element:
            # Extract the price value
            price = price_element.text.strip()
            return price
        else:
            print("Element with data-test='price' not found on the website.")
            return None
    else:
        print(f"Failed to retrieve the HTML content. Status code: {response.status_code}")
        return None

# Specify the URL of the website
website_url = "https://www.amazon.ca/gcx/Holiday-Toy-List/gfhz/events/?_encoding=UTF8&canBeEGifted=false&canBeGiftWrapped=false&categoryId=holiday-toy-list&content-id=amzn1.sym.28e3c471-8ef7-4cbb-8b9b-e3b4035a19b5&getItByTomorrow=false&isLimitedTimeOffer=false&isPrime=false&pd_rd_r=a59b0db4-ba3e-44eb-a66d-ce5bb0f15638&pd_rd_w=UjJYg&pd_rd_wg=mpfJp&pf_rd_p=28e3c471-8ef7-4cbb-8b9b-e3b4035a19b5&pf_rd_r=14AN1A6CQ2H94DKAJDJ2&ref_=Hol23_primary_vis_nav&scrollState=eyJpdGVtSW5kZXgiOjIsInNjcm9sbE9mZnNldCI6LTUwNC4yNjY2NjI1OTc2NTYyNX0%3D&sectionManagerState=eyJzZWN0aW9uVHlwZUVuZEluZGV4Ijp7ImFtYWJvdCI6MH19"

# Call the function and print the result
price_value = scrape_price_from_website(website_url)

if price_value:
    print(f"The price value is: {price_value}")
else:
    print("Failed to extract the price value.")
```

```
def extract_book_data(tag):
  title = tag.find("h3").find("a")["title"]
  price = (tag.find("p", attrs={"class": "price_color"}).text)  #regex101.com
  rating = tag.find("p", attrs={"class": "star-rating"})["class"][-1]

  return {
      "title" : title,
      "price": float(price[1:]),
      "rating": rating
  }
```

```
def get_job_for(lat=None, lng=None, results=5):
  if lat is None or lng is None:
    raise ValueError("Both lat and lng must be provided.")

  url = "https://api.higherme.com/classic/jobs"

  querystring = {"page":"1","includes":"location,location.company,location.externalServiceReferences","limit":results,"filters[brand.id]":"58bd9e7f472bd","filters[lat]":lat,"filters[lng]":lng,"filters[distance]":"6.25","sort[distance]":"asc"}

  headers = {
      "User-Agent": "Mozilla/5.0 (X11; Linux x86_64; rv:120.0) Gecko/20100101 Firefox/120.0",
      "Accept": "application/json, text/plain, */*",
      "Accept-Language": "en-CA,en-US;q=0.7,en;q=0.3",
      "Accept-Encoding": "gzip, deflate, br",
      "HigherMe-Client-Version": "2023.12.20_39.1",
      "Origin": "https://app.higherme.com",
      "DNT": "1",
      "Sec-GPC": "1",
      "Connection": "keep-alive",
      "Sec-Fetch-Dest": "empty",
      "Sec-Fetch-Mode": "cors",
      "Sec-Fetch-Site": "same-site",
      "TE": "trailers"
  }

  response = requests.request("GET", url, headers=headers, params=querystring)
  print(response.status_code)
  return response.json()
```
